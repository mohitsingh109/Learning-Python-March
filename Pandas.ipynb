```python
# ETL
# Merge & Concat

# merge() --> is used to combine two dataframes based on a common column or index.
#pd.merge(left, right, how='inner', on=None, left_on=None,right_on=None)
# left, right --> two dataframes to be merged
# on --> column name to merge on
# how --> type of merge to be performed (inner, outer, left, right)
# how='inner' --> only the rows with matching values in both dataframes will be included in the result
# left_on --> column name in left dataframe to merge on
# right_on --> column name in right dataframe to merge on
```

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5, 6],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Mohit', 'Karan'],
})
df1
```

```python
df2 = pd.DataFrame({
    'ID': [1, 2, 3, 44],
    'Score': [25, 30, 35, 40],
})
df2
```

```python
# Inner Join
# Only row with matching key (on='ID') are included
result = pd.merge(df1, df2, on='ID', how='inner')
result
```

```python
# Left Join
# All rows from the left df
result = pd.merge(df1, df2, on='ID', how='left')
result
```

```python
# Right Join
# All rows from the right df is included
result = pd.merge(df1, df2, on='ID', how='right')
result
```

```python
# Outer Join
# All rows from both df are included
# Outer = left + right
result = pd.merge(df1, df2, on='ID', how='outer')
result
```

```python
# Merge on multiple keys
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Subject': ['Science', 'Math', 'History'],
    'Score': [2, 4, 6]
})
df1
```

```python
df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Subject': ['Math', 'History', 'English'],
    'Grade': ['A', 'B', 'C']
})
df2
```

```python
result = pd.merge(df1, df2, on=['ID', 'Subject'], how='inner')
result
```

```python
# Merge with different column names
df1 = pd.DataFrame({
    'Student_ID': [1, 2, 3],
    'Subject': ['Science', 'Math', 'History'],
    'Score': [2, 4, 6]
})
df1
```

```python
df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Subject': ['Math', 'History', 'English'],
    'Grade': ['A', 'B', 'C']
})
df2
```

```python
result = pd.merge(df1, df2, left_on=['Student_ID', 'Subject'], right_on=['ID', 'Subject'], how='inner')
result
```
