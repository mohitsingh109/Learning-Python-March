```python
import pandas as pd
```

```python
# Dict
data = {
    "Name": ['Mohit', 'Karan', 'Aman'],
    "Age": [20, 30, 40],
    "Salary": [5, 10, 15],
    "Address": ["M1", None ,"A1"]
}

df = pd.DataFrame(data)
df
```

```python
df = pd.DataFrame(data, index=['E1', 'E2', 'E3'])
df
```

```python
# List
data = [
    {"Name": "Mohit", "Age": 20, "Salary": 10, "Address": "M1"},
    {"Name": "Karan", "Age": 30, "Salary": 20, "PhoneNumber": 20304034},
    {"Name": "Aman", "Age": 35, "Salary": 25, "Address": "A1"}
]
df = pd.DataFrame(data)
df
```

```python
# Convert Numpy array to DataFrame
import numpy as np
data = np.array([[25, 1000], [30, 3500], [40, 4500]])
df = pd.DataFrame(data, columns=['Age', 'Salary'], index=['Mohit', 'Karan', 'Aman'])
df
```

```python
# Empty DataFrame with columns
df = pd.DataFrame(columns=['Name', 'Age', 'Salary'])
df.loc[0] = ['Mohit', 20, 1000]
df.loc[1] = ['Karan', 30, 3500]
df.loc[2] = ['Aman', 40, 4500]
df
```

```python
# Sales Data Analysis

data = {
    'Month': ['Jan', 'Feb', 'Mar'],
    'Region A': [1000, 2000, 3000],
    'Region B': [4050, 5500, 6100],
    'Region C': [720, 800, 9500]
}
df = pd.DataFrame(data)
df
```

```python
# Adding new column called Total Sales
df['Total Sales'] = df['Region A'] + df['Region B'] + df['Region C']
df
```

```python
# Add avg sales
df["Avg Sales"] = df["Total Sales"] / 3
df
```

```python
# How to delete and insert data in Pandas DataFrame
```

```python
# drop() method is used to delete data from DataFrame
# label ==> column name, index name
# axis = 0 row (default), 1 column
# inplace = True/False --> default is False, if True then changes will be made in original DataFrame
# error = 'ignore'/'raise' --> default is 'raise', if 'ignore' then no error will be raised if label not found
```

```python
data = [
    {"Name": "Mohit", "Age": 20, "Salary": 10, "Address": "M1"},
    {"Name": "Karan", "Age": 30, "Salary": 20, "PhoneNumber": 20304034},
    {"Name": "Aman", "Age": 35, "Salary": 25, "Address": "A1"}
]
df = pd.DataFrame(data)
df
```

```python
# I want to delete the Karan row
df_dropped = df.drop(labels=1, axis=0)
df_dropped
```

```python
# Pass multiple index to delete row
df_dropped = df.drop(labels=[0, 2], axis=0)
df_dropped
```

```python
# Delete a column
df_dropped = df.drop(labels='Salary', axis=1)
df_dropped
```

```python
df_dropped = df.drop(labels=['Salary', 'Address'], axis=1)
df_dropped
```

```python
# Delete using del keyword of python
del df['Salary']
df
```

```python
df_dropped = df.drop(labels=['Salary', 'Address'], axis=1, errors='ignore')
df_dropped
```
