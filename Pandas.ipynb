```python
import pandas as pd
```

```python
# Dict
data = {
    "Name": ['Mohit', 'Karan', 'Aman'],
    "Age": [20, 30, 40],
    "Salary": [5, 10, 15],
    "Address": ["M1", None ,"A1"]
}

df = pd.DataFrame(data)
df
```

```python
df = pd.DataFrame(data, index=['E1', 'E2', 'E3'])
df
```

```python
# List
data = [
    {"Name": "Mohit", "Age": 20, "Salary": 10, "Address": "M1"},
    {"Name": "Karan", "Age": 30, "Salary": 20, "PhoneNumber": 20304034},
    {"Name": "Aman", "Age": 35, "Salary": 25, "Address": "A1"}
]
df = pd.DataFrame(data)
df
```

```python
# Convert Numpy array to DataFrame
import numpy as np
data = np.array([[25, 1000], [30, 3500], [40, 4500]])
df = pd.DataFrame(data, columns=['Age', 'Salary'], index=['Mohit', 'Karan', 'Aman'])
df
```

```python
# Empty DataFrame with columns
df = pd.DataFrame(columns=['Name', 'Age', 'Salary'])
df.loc[0] = ['Mohit', 20, 1000]
df.loc[1] = ['Karan', 30, 3500]
df.loc[2] = ['Aman', 40, 4500]
df
```

```python
# Sales Data Analysis

data = {
    'Month': ['Jan', 'Feb', 'Mar'],
    'Region A': [1000, 2000, 3000],
    'Region B': [4050, 5500, 6100],
    'Region C': [720, 800, 9500]
}
df = pd.DataFrame(data)
df
```

```python
# Adding new column called Total Sales
df['Total Sales'] = df['Region A'] + df['Region B'] + df['Region C']
df
```

```python
# Add avg sales
df["Avg Sales"] = df["Total Sales"] / 3
df
```

```python
# How to delete and insert data in Pandas DataFrame
```

```python
# drop() method is used to delete data from DataFrame
# label ==> column name, index name
# axis = 0 row (default), 1 column
# inplace = True/False --> default is False, if True then changes will be made in original DataFrame
# error = 'ignore'/'raise' --> default is 'raise', if 'ignore' then no error will be raised if label not found
```

```python
data = [
    {"Name": "Mohit", "Age": 20, "Salary": 10, "Address": "M1"},
    {"Name": "Karan", "Age": 30, "Salary": 20, "PhoneNumber": 20304034},
    {"Name": "Aman", "Age": 35, "Salary": 25, "Address": "A1"}
]
df = pd.DataFrame(data)
df
```

```python
# I want to delete the Karan row
df_dropped = df.drop(labels=1, axis=0)
df_dropped
```

```python
# Pass multiple index to delete row
df_dropped = df.drop(labels=[0, 2], axis=0)
df_dropped
```

```python
# Delete a column
df_dropped = df.drop(labels='Salary', axis=1)
df_dropped
```

```python
df_dropped = df.drop(labels=['Salary', 'Address'], axis=1)
df_dropped
```

```python
# Delete using del keyword of python
del df['Salary']
df
```

```python
df_dropped = df.drop(labels=['Salary', 'Address'], axis=1, errors='ignore')
df_dropped
```

```python
# Insert
# insert()
# assign()
# direct assignment
```

```python
# insert() --> Use to add column
# df.insert(loc, column, value, allow_duplicates=False)
# Loc --> index of column where new column will be inserted
# column --> name of new column
# value --> value of new column
# allow_duplicates --> default is False, if True then duplicate column name will be allowed
```

```python
data = [
    {"Name": "Mohit", "Age": 20, "Salary": 10},
    {"Name": "Karan", "Age": 30, "Salary": 20},
    {"Name": "Aman", "Age": 35, "Salary": 25}
]
df = pd.DataFrame(data)
df
```

```python
df.insert(loc=1, column='Address', value=['M1', 'A1', 'A1'], allow_duplicates=True)
df
```

```python
df['Score'] = [90, 80, 70]
df
```

```python
# assign() -> create new dataframe with the added column

new_df = df.assign(PhoneNumber=[1234567890, 9876543210, 1122334455])
new_df
```

```python
df
```

```python
# How to insert a row in Pandas DataFrame
# loc or iloc can be used to insert a row in Pandas DataFrame
```

```python
data = [
    {"Name": "Mohit", "Age": 20, "Salary": 10},
    {"Name": "Karan", "Age": 30, "Salary": 20},
    {"Name": "Aman", "Age": 35, "Salary": 25}
]
df = pd.DataFrame(data)
df
```

```python
new_row = {"Name": "Rohit", "Age": 28, "Salary": 15}
df.loc[3] = new_row
```

```python
df
```

```python
new_row = {"Name": "Arjun", "Age": 28, "Salary": 34}
df.iloc[0] = new_row # It will insert the data at position 0 and shift the rest of the rows down
```

```python
df
```

```python
# concat() -> is a pandas function used to add two dataframes together. It can be used to add rows or columns to a dataframe.
new_row = [{"Name": "BlaBla", "Age": 444, "Salary": 444565}]
new_row_df = pd.DataFrame(new_row)
new_row_df
```

```python
result_df = pd.concat([df, new_row_df], ignore_index=True)
result_df
```
